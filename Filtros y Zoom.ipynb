```python
img = Image.open('/content/drive/MyDrive/Maestría en Ciencias de la Computación/Visión Computacional/food.jpeg')
```

```python


# Calcular el nuevo tamaño de la imagen
new_size = (img.width * 10, img.height * 10)

# Hacer el zoom con interpolación bilineal
zoomed_img = img.resize(new_size, resample=Image.BILINEAR)

width, height = img.size
# Crear la figura con el mismo tamaño que la imagen
fig = plt.figure(figsize=(width/100, height/100))
# Mostrar la imagen
plt.imshow(img)
# Desactivar los ejes
plt.axis('off')
# Mostrar la figura
plt.show()

width, height = zoomed_img.size
# Crear la figura con el mismo tamaño que la imagen
fig = plt.figure(figsize=(width/100, height/100))
# Mostrar la imagen
plt.imshow(zoomed_img)
# Desactivar los ejes
plt.axis('off')
# Mostrar la figura
plt.show()
```

```python
new_size = (img.width * 10, img.height * 10)

# Hacer el zoom con interpolación bicúbica
zoomed_img = img.resize(new_size, resample=Image.BICUBIC)

width, height = img.size
# Crear la figura con el mismo tamaño que la imagen
fig = plt.figure(figsize=(width/100, height/100))
# Mostrar la imagen
plt.imshow(img)
# Desactivar los ejes
plt.axis('off')
# Mostrar la figura
plt.show()

width, height = zoomed_img.size
# Crear la figura con el mismo tamaño que la imagen
fig = plt.figure(figsize=(width/100, height/100))
# Mostrar la imagen
plt.imshow(zoomed_img)
# Desactivar los ejes
plt.axis('off')
# Mostrar la figura
plt.show()
```
